```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Now load the extracted credit_score.csv into a DataFrame
df = pd.read_csv('/content/credit_score[1].csv')
display(df.head())
```

```python
df = df.drop(columns=["ID","Customer_ID","Name","SSN","Type_of_Loan","Credit_History_Age"], errors='ignore')
display(df.head())
```

```python
df["Age"] = df["Age"].astype(str).str.replace("_","")
df["Age"] = df["Age"].astype(int)
df["Occupation"] = df["Occupation"].replace("_____",np.nan)
df["Annual_Income"] = df["Annual_Income"].astype(str).str.replace("_","")
df["Annual_Income"] = df["Annual_Income"].astype(float)
df["Num_of_Loan"] = df["Num_of_Loan"].astype(str).str.replace("_","")
df["Num_of_Loan"] = df["Num_of_Loan"].astype(int)
df["Num_of_Delayed_Payment"] = df["Num_of_Delayed_Payment"].astype(str).str.replace("_","")
df["Num_of_Delayed_Payment"] = df["Num_of_Delayed_Payment"].astype(float)
df["Changed_Credit_Limit"] = df["Changed_Credit_Limit"].astype(str).str.replace("_","").replace("_____",np.nan).replace('', np.nan)
df["Changed_Credit_Limit"] = df["Changed_Credit_Limit"].astype(float)
df["Credit_Mix"] = df["Credit_Mix"].replace("_",np.nan)
df["Credit_Mix"] = df["Credit_Mix"].replace(["Standard","Good","Bad"],[1,2,0])
df["Outstanding_Debt"] = df["Outstanding_Debt"].astype(str).str.replace("_","").replace("_____",np.nan)
df["Outstanding_Debt"] = df["Outstanding_Debt"].astype(float)
df["Payment_of_Min_Amount"] = df["Payment_of_Min_Amount"].astype(str).str.upper() # Standardize case
df["Payment_of_Min_Amount"] = df["Payment_of_Min_Amount"].replace("NM","NO") # Replace NM with NO
df["Payment_of_Min_Amount"] = df["Payment_of_Min_Amount"].replace(["YES","NO"],[1,0]) # Map YES/NO to 1/0
df["Payment_of_Min_Amount"] = pd.to_numeric(df["Payment_of_Min_Amount"], errors='coerce') # Convert to numeric, coerce errors
df["Amount_invested_monthly"] = df["Amount_invested_monthly"].astype(str).str.replace("_","")
df["Amount_invested_monthly"] = df["Amount_invested_monthly"].astype(float)
df["Payment_Behaviour"] = df["Payment_Behaviour"].replace("!@9#%8",np.nan)
df["Monthly_Balance"] = df["Monthly_Balance"].astype(str).str.replace("_","")
df["Monthly_Balance"] = df["Monthly_Balance"].astype(float)
df["Credit_Score"] = df["Credit_Score"].replace(["Poor","Standard","Good"],[0,1,2])
```

```python
df = df.fillna(method = "ffill")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Month"] = le.fit_transform(df["Month"])
df["Occupation"] = le.fit_transform(df["Occupation"])
df["Payment_Behaviour"] = le.fit_transform(df["Payment_Behaviour"])
df.info()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

col_list = []
for col in df.columns:
    if (df[col].dtype != "object") and (col != "Credit_Score"):
        col_list.append(col)

x = df[col_list]
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
vif_data["VIF"] = [variance_inflation_factor(x.values, i)
                  for i in range(len(x.columns))]
print(vif_data)
```

```python
x = df.drop(columns = "Credit_Score")
y = df["Credit_Score"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
pd.DataFrame({"actual_value":y_test,"predicted_value":y_pred})
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
accuracy_score(y_test,y_pred)
pd.DataFrame({"actual_value":y_test,"predicted_value":y_pred})
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

parameters = {"max_features": ["sqrt","log2"],
              "criterion": ["gini","entropy"],
              "max_depth": [2,3,5,10,50],
              "min_samples_split": [2,3,50,100],
              "min_samples_leaf": [1,5,8,10]
              }
grid_search = GridSearchCV(dt,parameters)
grid_search.fit(x_train,y_train)
dt = grid_search.best_estimator_
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
acc_dt = round(accuracy_score(y_test,y_pred) * 100,2)
print("accuracy of Decision Tree Model:",acc_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
accuracy_score(y_test,y_pred)
pd.DataFrame({"actual_value":y_test,"predicted_value":y_pred})
```
